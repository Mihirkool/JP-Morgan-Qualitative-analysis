```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from datetime import datetime, timedelta

# Load the dataset
df = pd.read_csv('Nat_Gas.csv')

# Convert the date column to datetime
df['Date'] = pd.to_datetime(df['Date'])

# Sort data just in case
df = df.sort_values('Date')

# Plot the data to visualize trends
plt.figure(figsize=(10,5))
plt.plot(df['Date'], df['Price'], marker='o')
plt.title('Natural Gas Prices Over Time')
plt.xlabel('Date')
plt.ylabel('Price ($)')
plt.grid(True)
plt.show()

# Convert dates to ordinal (for regression)
df['DateOrdinal'] = df['Date'].map(datetime.toordinal)

# Train Linear Regression model
X = df[['DateOrdinal']]
y = df['Price']
model = LinearRegression()
model.fit(X, y)

# Create a function to predict price for any date
def estimate_price(date_str):
    date_obj = datetime.strptime(date_str, '%Y-%m-%d')
    date_ord = date_obj.toordinal()
    predicted_price = model.predict([[date_ord]])
    return round(predicted_price[0], 2)

# Example usage:
print("Estimated price for 2023-07-15:", estimate_price("2023-07-15"))
print("Estimated price for 2025-08-01:", estimate_price("2025-08-01"))

# Extrapolate next 12 months
last_date = df['Date'].max()
future_dates = [last_date + pd.DateOffset(months=i) for i in range(1, 13)]
future_ordinals = [d.toordinal() for d in future_dates]
future_prices = model.predict(np.array(future_ordinals).reshape(-1, 1))

# Plot future predictions
plt.figure(figsize=(10,5))
plt.plot(df['Date'], df['Price'], label='Historical')
plt.plot(future_dates, future_prices, label='Predicted', linestyle='--', color='red')
plt.title('Natural Gas Prices with 1-Year Forecast')
plt.xlabel('Date')
plt.ylabel('Price ($)')
plt.legend()
plt.grid(True)
plt.show()
```
